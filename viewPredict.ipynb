```python
# 這邊測試不同的y
import math
def trend_log(x):
    x = int(x)
    if x <= 1:
        return 0
#     elif x < 1:
#         return math.log10(-1*x) * -1
    else:
        return math.log10(x)

def change48VToClass2(num):
    if num <= 1000:
        return "1000"
    elif num <= 5000:
        return "5000"
    elif num <= 10000:
        return "10000"
    elif num <= 50000:
        return "50000"
    elif num <= 100000:
        return "100k"
    elif num <= 500000:
        return "500k"
    elif num <= 1000000:
        return "1m"
    else:
        return ">1m"
    
# y_train.apply(trend_log)
def change48VToClass3(num):
    if num <= vsort[0]:
        return "10%"
    elif num <= vsort[1]:
        return "20%"
    elif num <= vsort[2]:
        return "30%"
    elif num <= vsort[3]:
        return "40%"
    elif num <= vsort[4]:
        return "50%"
    elif num <= vsort[5]:
        return "60%"
    elif num <= vsort[6]:
        return "70%"
    elif num <= vsort[7]:
        return "80%"
    elif num <= vsort[8]:
        return "90%"
    else:
        return "100%"

    # y_train.apply(trend_log)
def change48VToClass4(num):
    if num <= vsort[0]:
        return 0
    elif num <= vsort[1]:
        return 1
    elif num <= vsort[2]:
        return 2
    elif num <= vsort[3]:
        return 3
    elif num <= vsort[4]:
        return 4
    elif num <= vsort[5]:
        return 5
    elif num <= vsort[6]:
        return 6
    elif num <= vsort[7]:
        return 7
    elif num <= vsort[8]:
        return 8
    else:
        return 9
```

```python
def c_int(x):
    if x is None:
        return 0
    return int(x)

# data_df["sentiment"] = data_df["sentiment"].apply(c_int)
data_df["sentiment"] = data_df["sentiment"].apply(c_int).apply(change48VToClass4)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_df["content"], data_df["sentiment"], test_size=0.2)
```

```python
train_df = pd.DataFrame(columns = ["content", "views"])
train_df["content"] = x_train
train_df["views"] = y_train
```

```python
test_df = pd.DataFrame(columns = ["content", "views"])
test_df["content"] = x_test
test_df["views"] = y_test
# test_df
```

```python
# 讀取自己準備的字頻
with open("data/study/feq","r",encoding="utf-8") as fqv:
    feq = fqv.read()
feqList = feq.split("\n")
```

```python
# 這是直接抓文章裡的字頻
# from keras.preprocessing.text import Tokenizer
# token = Tokenizer(num_words=2000)
# token.fit_on_texts(train_df["content"])
# # 我省略了這裡的印出，讀者可以把註解秀出字典的樣子
# token.word_index
```

```python
from keras.preprocessing.text import Tokenizer
token = Tokenizer(num_words=2000)
token.fit_on_texts(feqList)
# 我省略了這裡的印出，讀者可以把註解秀出字典的樣子
# token.word_index
```

```python
x_train_seq = token.texts_to_sequences(train_df["content"])
x_test_seq = token.texts_to_sequences(test_df["content"])
pd.DataFrame(x_train_seq)
```

```python
from keras.preprocessing import sequence
x_train_pad = sequence.pad_sequences(x_train_seq, maxlen = 50)
x_test_pad = sequence.pad_sequences(x_test_seq, maxlen = 50)
pd.DataFrame(x_train_pad)
```

```python
x_train_pad[1]
```

```python
from keras.utils import np_utils 
y_train = train_df['views']
y_test = test_df['views']
y_train_cat = np_utils.to_categorical(y_train)
y_test_cat = np_utils.to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Flatten
from keras.layers.embeddings import Embedding
model = Sequential()
model.add(Embedding(output_dim=32, input_dim=2000, input_length=50))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.35))
# 注意一下, 因為我們是二元分類, 最後的激勵函數選擇 sigmoid
# sigmoid(正 + 負 =100%) softmax(類別全部 =100%)
model.add(Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

```python
import numpy as np
# 特別注意一下, 因為我們只是二元分類, 所以這裡的 loss 選擇 binary_crossentropy # mean_absolute_error mean_squared_error
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc'])

from keras.datasets import imdb
# y_train = train_df['views']
train_history = model.fit(x_train_pad, y_train_cat,batch_size = 100, epochs = 3, verbose = 2, validation_split = 0.2)
```

```python
# 正確率是 list 第二個元素
model.evaluate(x_test_pad, y_test_cat)
```
