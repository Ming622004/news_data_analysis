```python
import math
def trend_log(x):
    x = int(x)
    if abs(x) <= 1:
        return 0
    elif x < 1:
        return math.log10(-1*x) * -1
    else:
        return math.log10(x)

def change48VToClass2(num):
    if num <= 1000:
        return "1000"
    elif num <= 5000:
        return "5000"
    elif num <= 10000:
        return "10000"
    elif num <= 50000:
        return "50000"
    elif num <= 100000:
        return "100k"
    elif num <= 500000:
        return "500k"
    elif num <= 1000000:
        return "1m"
    else:
        return ">1m"

def change48VToClass3(num):
    if num <= 1000:
        return 0
    elif num <= 5000:
        return 1
    elif num <= 10000:
        return 2
    elif num <= 50000:
        return 3
    elif num <= 100000:
        return 4
    elif num <= 500000:
        return 5
    elif num <= 1000000:
        return 6
    else:
        return 7
# y_train.apply(trend_log)
```

```python
def c_int(x):
    return int(x)
    
data_df["sentiment"] = data_df["sentiment"].apply(c_int)
max_v = max(data_df["sentiment"])
min_v = min(data_df["sentiment"])
```

```python
def c_v_10(x):
    return ((x-min_v) / (max_v-min_v))
data_df["sentiment"] = data_df["sentiment"].apply(change48VToClass3)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_df["content"], data_df["sentiment"], test_size=0.2)
```

```python
train_df = pd.DataFrame(columns = ["content", "views"])
train_df["content"] = x_train
train_df["views"] = y_train
```

```python
test_df = pd.DataFrame(columns = ["content", "views"])
test_df["content"] = x_test
test_df["views"] = y_test
test_df
```

```python
with open("data/study/feq","r",encoding="utf-8") as fqv:
    feq = fqv.read()
feqList = feq.split("\n")
```

```python
from keras.preprocessing.text import Tokenizer
token = Tokenizer(num_words=2000)
token.fit_on_texts(train_df["content"])
# 我省略了這裡的印出，讀者可以把註解秀出字典的樣子
token.word_index
```

```python
from keras.preprocessing.text import Tokenizer
token = Tokenizer(num_words=2000)
token.fit_on_texts(feqList)
# 我省略了這裡的印出，讀者可以把註解秀出字典的樣子
token.word_index
```

```python
x_train_seq = token.texts_to_sequences(train_df["content"])
x_test_seq = token.texts_to_sequences(test_df["content"])
pd.DataFrame(x_train_seq)
```

```python
from keras.preprocessing import sequence
x_train_pad = sequence.pad_sequences(x_train_seq, maxlen = 100)
x_test_pad = sequence.pad_sequences(x_test_seq, maxlen = 100)
pd.DataFrame(x_train_pad)
```

```python
x_train_pad[1]
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Flatten
from keras.layers.embeddings import Embedding
model = Sequential()
model.add(Embedding(output_dim=32, input_dim=2000, input_length=100))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.35))
# 注意一下, 因為我們是二元分類, 最後的激勵函數選擇 sigmoid
# sigmoid(正 + 負 =100%) softmax(類別全部 =100%)
model.add(Dense(units=8, activation='softmax'))
```

```python
 model.summary()
```

```python
import numpy as np
# 特別注意一下, 因為我們只是二元分類, 所以這裡的 loss 選擇 binary_crossentropy # mean_absolute_error
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
from keras.utils import np_utils
from keras.datasets import imdb
y_train = train_df['views']
y_train_cat = np_utils.to_categorical(y_train)
y_train_cat
```

```python
train_history = model.fit(x_train_pad, y_train_cat,batch_size = 100, epochs = 3, verbose = 2, validation_split = 0.2)
```

```python
y_test = test_df['views']
y_test_cat = np_utils.to_categorical(y_test)
# 正確率是 list 第二個元素
model.evaluate(x_test_pad, y_test_cat)
```
